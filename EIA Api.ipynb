The categories api is a menu explorer. It will get you around the datasets, you navigate by changing the "category_id" field. 

```python
import requests
import json

query = "475136"

#Query the api, query is the numerical value of whatever category you want into from. 
url_call = "http://api.eia.gov/category/?api_key=" + key + "[&category_id=" + query +"][&out=json]"
r = requests.get(url_call) 
catResults = json.loads(r.text)

catResults
```

From the data you get from the categories api you can grap the 
    series_id 
value and use that to feed into the series api: 

```python
seriesId = "NG.RES_EPG0_R5302_NUS_BCF.A"

#Query the api, query is the numerical value of whatever category you want into from. 
url_call = "http://api.eia.gov/series/?api_key=" + key + "&series_id=" + seriesId 
r = requests.get(url_call) 
seriesResults = json.loads(r.text)

seriesResults
```

## General format for dealing with EIA API: 

1. Find the category you are most interested in (May be easiest with the [eia api browser](http://www.eia.gov/beta/api/qb.cfm))
2. Loop through all of the results and compile an array of the series_id's for that category. 
3. Set up another loop to run through those ids and query the api assembling the results into a nice array of dicts
4. Clean and profit.
