```python
from pandas import read_csv
import pandas as pd
data = read_csv("data/nuclearReactors.csv")
data.head()
```

```python
columns = ["plant", "location", "capacity","lonlats","displayLocations"]
df = pd.DataFrame(columns=columns)
df.plant = data["Plant Name, Unit Number"]
df.location = data["Location"]
df.capacity = data["Licensed MWt"]
df = df[df.location.notnull()]
df.head()
```

```python
import string
exclude = set(string.punctuation) #a list of punctuation

fixedLocs = []
dispLocs = []
for i in range(len(df)):
    string = df.ix[i].location
    where = string.index('(') #Get rid of the qualitative discriptions
    string = string[:where - 1]
    
    dispLocs.append(string)
    
    newString = ''.join(ch for ch in string if ch not in exclude) #Get rid of punctuation
    
    newString = newString.replace("   ","+") #Puts location into geolocationAPi friendly format 
    newString = newString.replace("  ","+")
    newString = newString.replace(" ","+")
    
    fixedLocs.append(newString)

fixedLocs
#dispLocs
```

```python
df.location = fixedLocs
df.displayLocations = dispLocs
df.head()
```
