### Data from http://www.nrel.gov/gis/data_solar.html
---

```python
from pandas import read_csv
import pandas as pd
data = read_csv("data/originalDatasets/solarLevelsCounty.csv")
data.head()
```

```python
df = pd.DataFrame(columns = ['address']) #initialize dataframe. 
df.head()
```

```python
df.address = data.County + "+" + data.State
df['annualAverage'] = data["Annual Average (kWh/m2/day)"]
data.to_csv('data/solarLevelsCounty.csv')
df.head()
```

```python
df = df[df.annualAverage != '-']
df.annualAverage = [eval(entry) for entry in df.annualAverage]
```

```python
df["lonlats"] = [geolocations(place) for place in df.address]
```

```python
df["lon"] = [entry[0] for entry in df.lonlats]
df["lat"] = [entry[1] for entry in df.lonlats]
df.head()
```
